```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Fix spaces in column names : 
df.columns = df.columns.str.strip()
df.columns
```

```python
df[['day','month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day','month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI',]].astype(float)
```

### Exploratory Data Analysis

```python
df_copy = df.drop(['day', 'month', 'year'], axis = 1,)
df_copy.head()
```

```python
df_copy[['Classes']].astype(str)
```

```python
# Encoding of the catogories

df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0, 1)
```

```python
df_copy
```

```python
# Plot density for all features
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#Percentage for pie-charts 
pie_percent = df_copy["Classes"].value_counts(normalize=True) * 100
```

```python
# Plotting the pie-charts
class_lables = ["Fire", "Not-fire"]

plt.figure(figsize=(10,8))
plt.pie(pie_percent,labels=class_lables,autopct='%1.1f%%')
plt.title("The Class Comparisons")
plt.show()
```

```python
df_corr = df_copy.corr() 
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
plt.figure(figsize=(14,12))
```

```python
# Box Plots

sns.boxplot(df_copy['FWI'], color='skyblue')
```
